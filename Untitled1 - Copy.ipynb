#### testing stationarity

```python

from statsmodels.tsa.stattools import adfuller
```

```python
# HO: it is non stationary
# H1: it is stationary

def adfuller_test(y):

    result = adfuller(y)
    
    labels = ['ADF Test Statistics', 'p-value', '#Lags used', 'Number of observations used']
    
    for value,label in zip(result, labels):
       
         print(label+":"+str(value))
    
    if result[1]<= 0.05:
       
         print('strong evidence against the null hypothesis(HO), rejecting the null hypothesis. data is stationary')
   
    else:
        
        print('weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary')
```

```python
adfuller_test(data_train['y'])
```

##### if data is not stationary , 'differencing' used to make it stationary. in this we shift 1 position down

```python
data_train.loc[:,('y first difference')] = data_train.loc[:, ('y')]-data_train.loc[:, ('y')].shift(1)
```

```python
data_train.head()
```

```python
data_train.loc[:, 'y first difference'] = data_train.loc[:, 'y']-data_train.loc[:, 'y'].shift(12)# data has seasonal difference (1year=12 months)
```

```python
data_train.head()
```

```python
data_train.isna().sum()
```

```python
# again testing dickey fuller test

adfuller_test(data_train['y first difference'].dropna())
```

##### now we have stationary data

### Forecasting a Time Series

### ARIMA

```python
import statsmodels.api as sm
from statsmodels.tsa.arima.model import ARIMA
```

```python
#ACF and PACF plots:
from statsmodels.tsa.stattools import acf, pacf
```

```python
acf = acf(data_train.loc[:,'y first difference'].dropna(), nlags=20)
pacf = pacf(data_train.loc[:,'y first difference'].dropna(), nlags=20, method='ols')
```

```python
#Ploting ACF
fig = plt.figure(figsize = (14,6)) 
plt.plot(acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(data_train['y first difference'].dropna())),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(data_train['y first difference'].dropna())),linestyle='--',color='gray')
plt.title('Autocorrelation Function')
plt.show()
```

```python
# ploting PACF
fig = plt.figure(figsize = (14,6)) 
plt.plot(pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(data_train['y first difference'].dropna())),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(data_train['y first difference'].dropna())),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function')
plt.show()
```

```python
model = ARIMA(data_train['y first difference'].dropna(), order=(2, 1, 2))
result = model.fit()
```

```python
result.summary()
```

```python
result.plot_diagnostics(figsize = (16,8))
plt.show()
```

```python
pred = result.get_predictions(start = pd.to_datetime('2020-09-14'), dynamic = False, end = pd.to_datetime('2020-10-04'))
```

```python
data_train['forecast1'] = result.predict(start = 896, end = 918, dynamic = True)
data_train[['y', 'forecast1']].plot(figsize = (16,8))
```
